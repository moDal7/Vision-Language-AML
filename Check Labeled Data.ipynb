### Libraries

```python
import json
import os
from PIL import Image
import matplotlib.pyplot as plt
```

### Visualize JSON labeled file

```python
#modify the file to be a working json file
with open('data/clip_labels.txt', 'r') as file:
    text = file.readlines()[0]
text = text[:-1] #remove ,
text = '[' + text #add [
text = text.replace('\'', '\"') #replace ' with "
text = text + ']'#add ]
with open('data/clip_labels.json', 'w') as file:
    file.write(text)
with open('data/clip_labels.json', 'r') as file:
    js_table = json.load(file)
os.remove('data/clip_labels.json')
for elem in js_table:
    print(elem)
```
